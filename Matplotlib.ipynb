```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
'''
La ventaja que tiene tambin subplots es poder acceder a los axes
o a las "graficas" a traves de indices.
'''
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
'''
Siendo tambin una ventaja, podemos definir un array
con una tupla. as pudiendo acceder al codigo de forma
orientada objetos. "Siendo un tipo de Clases"
'''
```

```python
fig, axes = plt.subplots(2,4)
#El primer valor dentro del parentesis contiene el numero de filas
#Mientras que el segundo valor es para las columnas
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), 'y')
axes[0,2].plot(x, np.tan(x), 'r')
axes[0,3].plot(x, np.cos(x)**2, 'g')
'''
Aqu estamos utilizando numpy con sus integraciones matematicas
adems estamos utilizando el metodo orientado a objetos
para poder grficar multiples funciones. Pudiendo modificar desde el color
la medida, el posicionamiento y los parametros.
'''
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
#Otra manera es realizando el array en forma de tupla anticipadamente
#Usando el metodo "Clases"
ax1.plot(x, np.cos(x), 'b.')
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x)**2, 'g')
ax5.plot(x, np.cos(x)+2,'-r')
ax6.plot(x, np.sin(x), 'r:')
ax7.plot(x, np.tan(x), '-yD')
ax8.plot(x, np.cos(x)**2, '-g.')
fig.tight_layout()
#Cambiando propiedad de grficas
'''
En esta prctica se junta los datos y genera una carga cognitiva negativa.
por lo tanto modificando slo la "fig" podemos acceder a otro tipo de
visualizacin.
'''
```

```python
x = np.linspace(0,5,11)
y= np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
#Utilizando 'figsize=(x,y))' podemos cambiar el tamao de visualizacin

axes[0].plot(x,y, label='$sin(x)$')
#Si agregamos '$math$ agregaremos una notacin matematica'
axes[0].legend()
#Agregamos una etiqueta para la grafica usando ".legend()"

axes[0].set_title('Relacin X - Y')
#Podemos utilizar Set para crear ttulos en nuestras graficas accediendo con su indice

axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
#Podemos tambin con set agregar descripciones, igual de forma de ndice.

#Ejemplo:
axes[1].plot(y,x)
axes[1].set_title('Relacin Y - X')
axes[1].set_xlabel('Etiqueta de X')
axes[1].set_ylabel('Etiqueta de Y')
fig.tight_layout()
#Esta propiedad nos permite ajustar de una mejor manera nuestra grfica.
```

```python
#Un ejemplo si no fuera orientado a objetos.
plt.plot(x,y,label='sin(x)')
plt.title('Este es un ttulo')
plt.xlabel('x')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(0.2,0.2))
```

```python
import matplotlib.pyplot as plt
import numpy as np
'''
Colores y estilos de Matplotlib
'''
```

```python
x = np.linspace(0,5,11)
#Set de Datos
```

```python
print(plt.style.available)
#Aqu nos muestra todos los estilos predefinidos
```

```python
# plt.style.use('bmh')
# plt.style.use('Solarize_Light2')
plt.style.use('dark_background')
'''
De esta manera podemos usar un estilo llamandolo
de forma tipo Clases. "plt.style.use('name')"
'''
```

```python
fig, ax = plt.subplots(figsize=(8,8))
#Comenzamos modelando nuestra grfica.
ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.')
ax.plot(x, x+4, 'g:')
ax.plot(x, x+5, '-cD')
#Aqu llamamos a nuestros valores con estilos diferentes.
'''
Ests son modificaciones tipo "Matlab" 
https://matplotlib.org/3.5.1/tutorials/colors/colors.html
estn todas las referencias de colores.
'''
```

```python
# Otra manerra es con Pylot, osea de forma no orientada a objetos.
fig, ax = plt.subplots(figsize=(8,8))
# Comenzamos modelando nuestra grfica.

ax.plot(x, x+1,color = 'green', linestyle='--')
# Con "linestyle='n'" podemos definir el tipo de lnea

ax.plot(x, x+2,color = 'blue', marker='P', markersize=10, markerfacecolor='#CCFFCC')
#Con "marker='n'" podemos usar marcadores en la grfica
#Con "markersize=n" cambiamos el tamao
#Con "markerfacecolor='n'" su color.

ax.plot(x, x+3,color = '#66FF66', linewidth=8)
# con "linewidth=n" podemos definir el grosor

ax.plot(x, x+4,color = '#CCFFCC')

ax.plot(x, x+5,color = 'red', alpha=0.8)
# Con "alpha=n" podemos definir la opacidad
```

'''
Barplot "Grficas de barras"
El grfico de barras nos permite graficar variables categricas, 
es decir, variables de texto, lo que es muy importante 
en el mundo de la ciencia de datos y Matplotlib ofrece ciertas 
caractersticas que nos facilita la vida en cuanto a graficar este tipo de variables.

'''

```python
plt.style.use('seaborn-dark')
#Definimos un estilo de grfica
```

```python
countries = ['INDIA', 'JAPON', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000,800,900,1000,300]
# Definimos nuestro Data set
```

```python
plt.bar(countries, population)
# llamamos a la grfica con "plt.bar"
plt.show()
# Visualizacin de la grfica
```

```python
plt.bar(countries, population, width=0.5)
#Con "width=n" definimos el grosor de nuestras barras
plt.show()
```

```python
plt.bar(countries, population, width=0.5, color=['cyan','blue'])
#Se puede agregar un color definido, o poner multiples colores
plt.show()
'''
Podemos agregar ms colores 
buscando paletas en: https://coolors.co/palettes/trending
'''
```

```python
plt.bar(countries, population, width=0.5, color=['#264653','#2A9D8F', '#E9C46A', '#F4A261', '#E76F51'])
plt.xticks(np.arange(5),('India','Japon','Mxico','Colombia','Alemania'),rotation=45)
#Aqu lo que hicimos con "plt.xticks" es acceder a la configuracin del eje 'x'
#Hacemos uso de np.arange(n) para que nos haga una lista con las posiciones que ya tenemos.
plt.show()
```

```python
plt.barh(countries, population, color=['#264653','#2A9D8F', '#E9C46A', '#F4A261', '#E76F51'])
#Con la caracteristica de "plt.barh" podemos hacer nuestras barras horizontales
plt.title('Poblacin de paises ejemplo:')
plt.show()
```

```python
#Otro ejemplo
#Import
import matplotlib.pyplot as plt
import numpy as np
#Data
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population =[1000,800,900,1000,300]
#Set style
plt.style.use('dark_background')
#Set bar_plot
plt.bar(country,population,width=0.5,color=['#FFBE0B','#FB5607','#FF006E','#8338EC', '#3A86FF'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
#Set context
plt.title('World Population Country')
plt.xlabel('Country')
plt.ylabel('Population',rotation=90)
#Show plot
plt.show()
```
